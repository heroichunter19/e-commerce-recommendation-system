```python
bookrating = bookrating[bookrating['userid'].isin(y)]
```

```python
#merge bookrating with data
bookrating_with_data = bookrating.merge(data,on='ISBN')
```

```python
bookrating_with_data.head()
```

```python
number_rating = bookrating_with_data.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.head()
```

```python
number_rating.rename(columns={'rating': 'num_rating'},inplace=True)
```

```python
number_rating.head()
```

```python
last_rating = bookrating_with_data.merge(number_rating,on='title')
```

```python
last_rating.head()
```

```python
last_rating =last_rating[last_rating['num_rating'] >= 50]
```

```python
last_rating.head()
```

```python
last_rating.drop_duplicates(['userid','title'],inplace=True)
```

```python
last_rating.isnull().sum()
```

```python
data_pivot_table= last_rating.pivot_table(columns='userid',index='title', values='rating')
```

```python
data_pivot_table
```

```python
data_pivot_table.fillna(0,inplace=True)
```

```python
data_pivot_table
```

```python
from scipy.sparse import csr_matrix
```

```python
data_sparse = csr_matrix(data_pivot_table)
```

```python
data_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(data_sparse)
```

```python
data_pivot_table.iloc[237,:]
```

```python
dis,sugg= model.kneighbors(data_pivot_table.iloc[237,:].values.reshape(1,-1), n_neighbors=6)
```

```python
def recom_book(nameOfbook):
    idOfbook = np.where(data_pivot_table.index == nameOfbook)[0][0]
    dis,sugg= model.kneighbors(data_pivot_table.iloc[idOfbook,:].values.reshape(1,-1), n_neighbors=6)

    for i in range(len(sugg)):
        allbooks = data_pivot_table.index[sugg[i]]
        for j in allbooks:
            if j ==nameOfbook:
                print(f"Searched Book'{nameOfbook}'\n")
                print("Suggestion Books for you: \n")
            else:
                print(j)
```
