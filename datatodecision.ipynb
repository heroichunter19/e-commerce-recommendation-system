```python
# rename columns name
data.rename(columns={
    "Book-Title":'title',
    "Book-Author": 'author',
    "Year-Of-Publication": 'pub-year',
    "Publisher": 'pub',
    "Image-URL-L": 'url'
},inplace=True)
```

```python
#again renameing userdata columns name
userdata.rename(columns={
    "User-ID":'userid',
    "Location": 'location',
    "Age": 'age'
},inplace=True)
```

```python
bookrating = pd.read_csv('BX-Book-Ratings.csv',sep=";", on_bad_lines='skip',encoding='latin-1')
```

```python
bookrating.head()
```

```python
bookrating.shape
```

```python
bookrating.rename(columns={
    "User-ID": 'userid',
    "Book-Rating": 'rating'
},inplace=True)
```

```python
print(data.shape,userdata.shape,bookrating.shape )
```

```python
bookrating['userid'].value_counts()
```

```python
x = bookrating['userid'].value_counts() > 200
```

```python
x
```

```python
x[x].shape
```

```python
y =x[x].index
```

```python
y
```

```python
bookrating = bookrating[bookrating['userid'].isin(y)]
```

```python
bookrating.head()
```

```python
#merge bookrating with data
bookrating_with_data = bookrating.merge(data,on='ISBN')
```

```python
bookrating_with_data.head()
```

```python
number_rating = bookrating_with_data.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.head()
```

```python
number_rating.rename(columns={'rating': 'num_rating'},inplace=True)
```

```python
number_rating.head()
```

```python
last_rating = bookrating_with_data.merge(number_rating,on='title')
```

```python
last_rating.head()
```

```python
last_rating =last_rating[last_rating['num_rating'] >= 50]
```

```python
last_rating.head()
```

```python
last_rating.drop_duplicates(['userid','title'],inplace=True)
```
